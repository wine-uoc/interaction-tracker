## BUILDING THE DATASET

```python
#create dataset
data = pd.DataFrame(columns=['timestamp', 'RSSI_A1', 'RSSI_A2', 'RSSI_A3', 'pos_x', 'pos_y'])

# Reading gathered data files
files_path = "/home/aaron/Trabajo/interaction-tracker-updated/PositionTracker/src/Data_Collection/Offline/3anchor_1movil/fingerprinting"
anchors_serName = ['LAUNCH1', 'LAUNCH2', 'LAUNCH3']

anchor1_files_list = sorted(list(filter(lambda x: ("detectedDevs" in x) and (anchors_serName[0] in x), os.listdir(files_path))))
anchor2_files_list = sorted(list(filter(lambda x: ("detectedDevs" in x) and (anchors_serName[1] in x), os.listdir(files_path))))
anchor3_files_list = sorted(list(filter(lambda x: ("detectedDevs" in x) and (anchors_serName[2] in x), os.listdir(files_path))))

#concatenate file's data to build the dataset
for fname_a1, fname_a2, fname_a3 in zip(anchor1_files_list, anchor2_files_list, anchor3_files_list):
    
    # Extracting X-Y Position
    pos_x = float(fname_a1.split('_')[3])
    pos_y = float(fname_a1.split('_')[5])
    
    #Build a dataframe for each 3 data files
    df_a1 = pd.read_json(files_path+'/'+fname_a1).drop(['launchpadId','devname'], axis=1).rename(columns={'RSSI':'RSSI_A1'})
    df_a2 = pd.read_json(files_path+'/'+fname_a2).drop(['launchpadId','devname'], axis=1).rename(columns={'RSSI':'RSSI_A2'})
    df_a3 = pd.read_json(files_path+'/'+fname_a3).drop(['launchpadId','devname'], axis=1).rename(columns={'RSSI':'RSSI_A3'})
    
    #Format timestamp as an integer
    df_a1['timestamp'] = df_a1['timestamp'].astype('int')
    df_a2['timestamp'] = df_a2['timestamp'].astype('int')
    df_a3['timestamp'] = df_a3['timestamp'].astype('int')
    
    #Merge data from 3 files into one dataframe with an inner join by timestamp
    df_pos = df_a1.merge(df_a2, on='timestamp', how='inner')
    df_pos = df_pos.merge(df_a3, on='timestamp', how='inner')
    
    #Insert positions data to dataframe
    df_pos.insert(loc=4, column='pos_x', value=np.repeat(pos_x, df_pos.shape[0]))
    df_pos.insert(loc=5, column='pos_y', value=np.repeat(pos_y, df_pos.shape[0]))
    
    print(df_pos)

    data = data.append(df_pos)

data.reset_index(drop=True, inplace=True)
```

## USING GAUSSIAN NAIVE BAYES MODEL

### BUILDING TRAIN & TEST SETS

```python
#split in train and test data
X_train, X_test, y_train, y_test = train_test_split(np.array(data['RSSI']).reshape(-1,1), np.array(data['dist_<_1m']).reshape(-1,1), test_size=data.shape[0]//3)
```

### APPLY GAUSSIAN NAIVE-BAYES

```python
#apply a gaussian naive-bayes
NB_model = GaussianNB()
NB_model.fit(X_train, y_train)

#using cross validation
NB_model_cv = GaussianNB()
scores = cross_val_score(NB_model_cv, X_train, y_train, cv=10)

#compute confusion matrix

y_test_pred = NB_model.predict(X_test)
print(confusion_matrix(y_test, y_test_pred))

print(f'cross validation score: mean->{scores.mean()} std->{scores.std()}\n')

#generate classification report
print("CLASSIFICATION REPORT")
y_test_pred = NB_model.predict(X_test)
print(classification_report(y_true=y_test, y_pred=y_test_pred))
```

## USING SVC

### BUILDING TRAIN & TEST SETS

```python
#split in train and test data
X_train, X_test, y_train, y_test = train_test_split(np.array(data['RSSI']).reshape(-1,1), np.array(data['dist_<_1m']).reshape(-1,1), test_size=data.shape[0]//3)
```

### SCALE TRAIN & TEST SETS SEPARATELY

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### APPLY SVC

```python
svc = SVC(class_weight='balanced')

kernels = ['linear', 'rbf', 'poly', 'sigmoid']
Cs = [0.0001, 0.001, 0.1, 0.5, 1, 2, 3, 10, 50 ,100, 1000]

trc = GridSearchCV(estimator=svc,
                   param_grid={
                       'C': Cs,
                       'kernel': kernels
                   },
                   scoring=['accuracy', 'recall_macro', 'f1_macro'],
                   cv=10,
                   n_jobs=-1,
                   return_train_score=True,
                   refit='f1_macro')

svc_model_10CV = trc.fit(X_train, y_train)

print(svc_model_10CV.best_score_)
print(svc_model_10CV.best_params_)
```

```python
# Try 'poly' kernel (the best for this case) with different degrees

svc = SVC(C=2, kernel='poly', class_weight='balanced')

degrees = [2,3,4,5,6,7,8,9,10]

trc = GridSearchCV(estimator=svc,
                   param_grid={
                       'degree': degrees
                   },
                   scoring=['accuracy', 'recall_macro', 'f1_macro'],
                   cv=5,
                   n_jobs=-1,
                   return_train_score=True,
                   refit='f1_macro')

svc_model_5CV = trc.fit(X_train, y_train)

print(svc_model_5CV.best_score_)
print(svc_model_5CV.best_params_)
```

```python
pd.DataFrame(svc_model_5CV.cv_results_).loc[:, [
    'mean_fit_time',
    'std_fit_time',
    'param_C',
    'param_kernel',
    'mean_test_accuracy',
    'std_test_accuracy',
    'mean_test_recall_macro',
    'std_test_recall_macro',
    'mean_test_f1_macro',
    'std_test_f1_macro',
]].sort_values(by='mean_test_f1_macro',ascending=False)
```

### RE-TRAIN WITH TEST DATA USING THE BEST MODEL

```python
svm = SVC(C=2, kernel='poly', degree=3, class_weight='balanced')
svm.fit(X_train,y_train)

scores = cross_val_score(svm, X_train, y_train, cv=10)
scores_recall = cross_val_score(svm, X_train, y_train, cv=10,scoring='recall_macro')
scores_f_score = cross_val_score(svm, X_train, y_train, cv=10,scoring='f1_macro')
[np.mean(scores),np.mean(scores_recall),np.mean(scores_f_score)]
```

```python
y_pred = svm.predict(X_test)

print(classification_report(y_test,y_pred))
```
